Cuplikan Artikel Unveiling the Practices of Research Publication Among Indonesian Academics

#Abstract

Research on poor-quality journals seems to be an inexhaustible study. The present study aims to understand the experiences of 3 Indonesian academics who published their scientific articles in disreputable journals. This phenomenology study was conducted within the framework of a qualitative approach. Data were collected by in-depth interviews academics. The study identified three key themes: Lack of Necessary Skills, Overestimation of Western Hegemony, and Acceptance of Fate. The findings indicated that some factors caused the academics to take shortcuts to get their articles published. The pressure to acquire the promotion to become a professor was the main factor influencing Indonesian academics to publish in these fast-turnaround journals. The findings highlight the need for academic institutions and policymakers to critically address the systemic pressures that drive scholars toward low-quality publishing outlets. Strengthening mentorship, providing training in academic writing, and reforming promotion policies to reward quality over quantity are essential to safeguard academic integrity. Furthermore, the study underscores the global implications of these issues, urging international academic bodies to collaborate in setting higher standards for research dissemination and evaluation practices.

#Introduction

Academic publication is not only a means of disseminating knowledge but also a metric for career progression, reputation, and institutional evaluation. However, the prestige associated with publication can sometimes push academics into situations where quantity is emphasized over quality. In contexts where institutional requirements or promotion criteria heavily weight publication counts, researchers may feel compelled to publish in journals with fast turnaround times and questionable reputations. This study explores these phenomena among Indonesian academics, focusing on how pressures related to promotion and career advancement intersect with perceptions of journal quality, writing skills, and institutional policies. The issue of publication in predatory journals has become increasingly concerning in Indonesia, particularly as higher education institutions continue to enforce publication as a requirement for academic promotion. Understanding the underlying factors behind this trend is crucial for developing effective strategies to promote ethical and sustainable research publication practices.

#Conclusion

The present study has revealed the involvement of three Indonesian academics in publishing their work in predatory journals, driven by intense pressure to meet publication requirements for academic promotion and the challenges of publishing in reputable, Scopus-indexed journals. This trend is largely attributed to three key factors: a lack of expertise in scholarly writing for international journals, a perception of bias and the likelihood of rejection from high-quality journals, and the influence of religious beliefs regarding fate in academic outcomes. Publishing in predatory journals has severe consequences for both individual academics and the broader academic community. On a personal level, it can damage an academic’s credibility and hinder career progression, as work published in disreputable outlets is often dismissed by peers and institutions. Additionally, the absence of rigorous peer review in these journals means that research may lack scientific validity, limiting its contribution to the field. This also harms the reputation of the institutions these academics represent, diluting the quality of academic output and diminishing their standing in global academia.

#Tabel pivot konsonan dan vokal dalam kata

```python
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

text = """
Research on poor-quality journals seems to be an inexhaustible study.
The present study aims to understand the experiences of three Indonesian academics
who published their scientific articles in disreputable journals.
This phenomenology study was conducted within the framework of a qualitative approach.
Data were collected by in-depth interviews.
The study identified three key themes: Lack of Necessary Skills, Overestimation of Western Hegemony, and Acceptance of Fate.
The findings indicated that some factors caused the academics to take shortcuts to get their articles published.
The pressure to acquire promotion to become a professor was the main factor influencing Indonesian academics to publish in these fast-turnaround journals.
The findings highlight the need for academic institutions and policymakers to critically address the systemic pressures that drive scholars toward low-quality publishing outlets.
Strengthening mentorship, providing training in academic writing, and reforming promotion policies to reward quality over quantity are essential to safeguard academic integrity.

Academic publication is not only a means of disseminating knowledge but also a metric for career progression, reputation, and institutional evaluation.
However, the prestige associated with publication can sometimes push academics into situations where quantity is emphasized over quality.
In contexts where institutional requirements or promotion criteria heavily weight publication counts, researchers may feel compelled to publish in journals with fast turnaround times and questionable reputations.
This study explores these phenomena among Indonesian academics, focusing on how pressures related to promotion and career advancement intersect with perceptions of journal quality, writing skills, and institutional policies.
The issue of publication in predatory journals has become increasingly concerning in Indonesia, particularly as higher education institutions continue to enforce publication as a requirement for academic promotion.
Understanding the underlying factors behind this trend is crucial for developing effective strategies to promote ethical and sustainable research publication practices.

The present study has revealed the involvement of three Indonesian academics in publishing their work in predatory journals, driven by intense pressure to meet publication requirements for academic promotion and the challenges of publishing in reputable, Scopus-indexed journals.
This trend is largely attributed to three key factors: a lack of expertise in scholarly writing for international journals, a perception of bias and the likelihood of rejection from high-quality journals, and the influence of religious beliefs regarding fate in academic outcomes.
Publishing in predatory journals has severe consequences for both individual academics and the broader academic community.
On a personal level, it can damage an academic’s credibility and hinder career progression, as work published in disreputable outlets is often dismissed by peers and institutions.
Additionally, the absence of rigorous peer review in these journals means that research may lack scientific validity, limiting its contribution to the field.
This also harms the reputation of the institutions these academics represent, diluting the quality of academic output and diminishing their standing in global academia.
Remedies include strengthening mentorship programs, improving training in scholarly communication, and revising institutional promotion policies to favor quality over quantity.
"""

def words_from_text(text):
    """Ambil semua kata dalam teks"""
    return re.findall(r"[a-zA-Z]+", text.lower())

def count_vowels_consonants(word):
    """Hitung jumlah vokal dan konsonan"""
    vowels = set("aeiou")
    v = sum(1 for c in word if c in vowels)
    c = sum(1 for c in word if c.isalpha() and c not in vowels)
    return c, v

words = words_from_text(text)
unique_words = sorted(set(words))


freq_counter = Counter()
unique_counter = Counter()

for w in words:
    c, v = count_vowels_consonants(w)
    freq_counter[(c, v)] += 1

for w in unique_words:
    c, v = count_vowels_consonants(w)
    unique_counter[(c, v)] += 1


def counter_to_df(counter):
    max_c = max(k[0] for k in counter.keys())
    max_v = max(k[1] for k in counter.keys())
    df = pd.DataFrame(0, index=range(max_v+1), columns=range(max_c+1))
    for (c, v), val in counter.items():
        df.loc[v, c] = val
    return df

df_freq = counter_to_df(freq_counter)
df_unique = counter_to_df(unique_counter)

print("=== Pivot Frekuensi Kata (termasuk duplikat) ===")
print(df_freq)
print("\n=== Pivot Kata Unik (tanpa duplikat) ===")
print(df_unique)

plt.figure(figsize=(8,6))
sns.heatmap(df_freq, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Heatmap Frekuensi Kata (Jumlah Vokal vs Konsonan)")
plt.xlabel("Jumlah Konsonan")
plt.ylabel("Jumlah Vokal")
plt.show()

plt.figure(figsize=(8,6))
sns.heatmap(df_unique, annot=True, fmt=".0f", cmap="OrRd")
plt.title("Heatmap Kata Unik (Jumlah Vokal vs Konsonan)")
plt.xlabel("Jumlah Konsonan")
plt.ylabel("Jumlah Vokal")
plt.show()
```

Dari tabel pivot ini kita dapat melihat bahwa :

~Kombinasi paling umum berada pada 3–5 konsonan dan 2–3 vokal, ciri khas kata ilmiah bahasa Inggris.

~Banyak kata pendek seperti in, to, of, on menyebabkan sel (1,1) tinggi pada frekuensi.

~Kata unik lebih banyak di kombinasi menengah (4–6 konsonan, 2–4 vokal) misalnya academic, quality, policy, integrity.

~Pola diagonal menunjukkan proporsi konsonan–vokal relatif stabil, menandakan keseimbangan fonetik umum pada bahasa ilmiah.

```python
freq_data = [(c, v, f) for (c, v), f in freq_counter.items()]
df_bubble = pd.DataFrame(freq_data, columns=['Consonants', 'Vowels', 'Frequency'])

plt.figure(figsize=(8,6))
sns.scatterplot(data=df_bubble, x="Consonants", y="Vowels", size="Frequency", sizes=(50, 500),
                alpha=0.6, hue="Frequency", palette="coolwarm", legend=False)
plt.title("Bubble Chart Distribusi Jumlah Konsonan dan Vokal")
plt.xlabel("Jumlah Konsonan")
plt.ylabel("Jumlah Vokal")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

bubble chart ini menggambarkan hubungan antara jumlah konsonan (sumbu X) dan jumlah vokal (sumbu Y) dalam setiap kata yang terdapat pada cuplikan artikel.

*Ukuran gelembung (bubble) menunjukkan frekuensi kemunculan kata dengan kombinasi tertentu.

*Warna merah menandakan titik dengan frekuensi tinggi, sedangkan warna biru muda menunjukkan frekuensi rendah.

*Dari grafik terlihat bahwa kata dengan 1–3 vokal dan 1–4 konsonan mendominasi teks, yang artinya struktur kata dalam artikel ini cenderung pendek dan mudah dibaca.

*Gelembung besar di area vokal rendah–konsonan rendah menandakan penggunaan banyak kata sederhana seperti "the", "of", "and", yang umum dalam tulisan akademik berbahasa Inggris.

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Konversi DataFrame menjadi grid 3D
X, Y = np.meshgrid(df_freq.columns, df_freq.index)
Z = df_freq.values

fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='k')
ax.set_title("Distribusi Frekuensi Kata (3D Surface)")
ax.set_xlabel("Jumlah Konsonan")
ax.set_ylabel("Jumlah Vokal")
ax.set_zlabel("Frekuensi")
plt.colorbar(surf, ax=ax, shrink=0.5)
plt.show()
```

#Jumlah kalimat dalam paragraf

```python
import matplotlib.pyplot as plt
import re

# Cuplikan artikel
text = """
Research on poor-quality journals seems to be an inexhaustible study.
The present study aims to understand the experiences of three Indonesian academics
who published their scientific articles in disreputable journals.
This phenomenology study was conducted within the framework of a qualitative approach.
Data were collected by in-depth interviews.
The study identified three key themes: Lack of Necessary Skills, Overestimation of Western Hegemony, and Acceptance of Fate.
The findings indicated that some factors caused the academics to take shortcuts to get their articles published.
The pressure to acquire promotion to become a professor was the main factor influencing Indonesian academics to publish in these fast-turnaround journals.
The findings highlight the need for academic institutions and policymakers to critically address the systemic pressures that drive scholars toward low-quality publishing outlets.
Strengthening mentorship, providing training in academic writing, and reforming promotion policies to reward quality over quantity are essential to safeguard academic integrity.

Academic publication is not only a means of disseminating knowledge but also a metric for career progression, reputation, and institutional evaluation.
However, the prestige associated with publication can sometimes push academics into situations where quantity is emphasized over quality.
In contexts where institutional requirements or promotion criteria heavily weight publication counts, researchers may feel compelled to publish in journals with fast turnaround times and questionable reputations.
This study explores these phenomena among Indonesian academics, focusing on how pressures related to promotion and career advancement intersect with perceptions of journal quality, writing skills, and institutional policies.
The issue of publication in predatory journals has become increasingly concerning in Indonesia, particularly as higher education institutions continue to enforce publication as a requirement for academic promotion.
Understanding the underlying factors behind this trend is crucial for developing effective strategies to promote ethical and sustainable research publication practices.

The present study has revealed the involvement of three Indonesian academics in publishing their work in predatory journals, driven by intense pressure to meet publication requirements for academic promotion and the challenges of publishing in reputable, Scopus-indexed journals.
This trend is largely attributed to three key factors: a lack of expertise in scholarly writing for international journals, a perception of bias and the likelihood of rejection from high-quality journals, and the influence of religious beliefs regarding fate in academic outcomes.
Publishing in predatory journals has severe consequences for both individual academics and the broader academic community.
On a personal level, it can damage an academic’s credibility and hinder career progression, as work published in disreputable outlets is often dismissed by peers and institutions.
Additionally, the absence of rigorous peer review in these journals means that research may lack scientific validity, limiting its contribution to the field.
This also harms the reputation of the institutions these academics represent, diluting the quality of academic output and diminishing their standing in global academia.
Remedies include strengthening mentorship programs, improving training in scholarly communication, and revising institutional promotion policies to favor quality over quantity.
"""

paragraphs = [p.strip() for p in text.strip().split('\n\n') if p.strip()]

# Hitung jumlah kalimat per paragraf
sentence_counts = []
for para in paragraphs:
    sentences = re.split(r'[.!?]+', para)
    sentences = [s.strip() for s in sentences if s.strip()]
    sentence_counts.append(len(sentences))

# Tampilkan hasil per paragraf
for i, count in enumerate(sentence_counts, 1):
    print(f"Paragraf {i} memiliki {count} kalimat.")

# Visualisasi histogram
plt.bar(range(1, len(sentence_counts)+1), sentence_counts, color='pink', edgecolor='black')
plt.xlabel("Nomor Paragraf")
plt.ylabel("Jumlah Kalimat")
plt.title("Jumlah Kalimat dalam Tiap Paragraf")
plt.xticks(range(1, len(sentence_counts)+1))
plt.show()
```

Pada output kita dapat melihat bahwa pada paragraf pertama memiliki jumlah kalimat terbanyak (9 kalimat), diikuti paragraf ketiga (7 kalimat) dan kedua (6 kalimat).

Visualisasi diatas memudahkan pembaca untuk langsung melihat perbandingan antar paragraf.

Tingginya batang pada paragraf pertama menunjukkan bagian pembuka artikel lebih panjang dan deskriptif, sedangkan paragraf berikutnya lebih ringkas dan fokus.

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=sentence_counts, color='pink', width=0.4)
plt.title("Sebaran Panjang Paragraf (Box Plot)")
plt.ylabel("Jumlah Kalimat per Paragraf")
plt.show()
```

Box plot menggambarkan sebaran jumlah kalimat per paragraf.

Terlihat bahwa nilai tengah (median) berada di sekitar 7 kalimat, dengan rentang (min–max) dari 6 hingga 9 kalimat.

Artinya, penulis memiliki gaya yang cukup konsisten dalam menulis paragraf, tanpa variasi ekstrem dalam panjang kalimat.

Tidak ada outlier, yang menunjukkan keseimbangan struktur antar paragraf.

```python
data = np.array(sentence_counts).reshape(1, -1)
plt.figure(figsize=(10,2))
sns.heatmap(data, annot=True, cmap="YlOrRd", cbar=True, fmt="d")
plt.title("Heatmap Jumlah Kalimat per Paragraf")
plt.xlabel("Paragraf ke-")
plt.ylabel("Jumlah Kalimat")
plt.show()
```

Heatmap memberikan tampilan visual “kepadatan panjang” tiap paragraf.

Warna yang lebih gelap menandakan paragraf dengan jumlah kalimat lebih banyak.

Pada hasil ini, paragraf pertama tampak paling gelap, menandakan panjang dan kompleksitasnya yang lebih tinggi.
Visualisasi ini efektif untuk melihat distribusi panjang paragraf secara cepat dan intuitif.
