Cuplikan Artikel Unveiling the Practices of Research Publication Among Indonesian Academics

#Abstract
Research on poor-quality journals seems to be an inexhaustible study. The present study aims to understand the experiences of 3 Indonesian academics who published their scientific articles in disreputable journals. This phenomenology study was conducted within the framework of a qualitative approach. Data were collected by in-depth interviews academics. The study identified three key themes: Lack of Necessary Skills, Overestimation of Western Hegemony, and Acceptance of Fate. The findings indicated that some factors caused the academics to take shortcuts to get their articles published. The pressure to acquire the promotion to become a professor was the main factor influencing Indonesian academics to publish in these fast-turnaround journals. The findings highlight the need for academic institutions and policymakers to critically address the systemic pressures that drive scholars toward low-quality publishing outlets. Strengthening mentorship, providing training in academic writing, and reforming promotion policies to reward quality over quantity are essential to safeguard academic integrity. Furthermore, the study underscores the global implications of these issues, urging international academic bodies to collaborate in setting higher standards for research dissemination and evaluation practices.

#Introduction

In today’s academic landscape, the expectation to publish in esteemed journals presents a significant challenge for scholars worldwide, as universities striving for high international rankings require academics to publish in reputable international journals (Gautam, 2020). Reputable journals are essential for fulfilling promotion criteria (Shrestha, 2021; Demir, 2018; Hedding, 2019), and publishing original research, review articles, or thought pieces in peer-reviewed outlets demonstrates expertise and contributes meaningfully to the field (Beall, 2012a). However, due to immense publication pressure, many scholars resort to submitting their work to international journals despite high fees (Raitskaya, 2020; Linacre et al., 2019), often falling victim to disreputable journals—publications that falsely present themselves as legitimate but engage in unethical practices such as dishonest editorial boards, improper peer review, and hidden charges (Nnodim et al., 2023; Grudniewicz et al., 2019; Elmore et al., 2020). These journals exploit researchers’ urgency to publish, leading to financial losses and damaged reputations (Beall, 2015; Cobey et al., 2018; Moher et al., 2017). Indonesia ranks second globally in publishing within poor-quality journals (Macháček et al., 2021), driven by promotion pressure, limited awareness, and the ease of online publishing (Xia, 2015), prompting universities and policymakers to implement awareness campaigns, blacklists, and quality-oriented publication guidelines (Arthur et al., 2021). Studies show that lack of mentorship and research experience also contribute to this issue, as inexperienced authors may struggle to assess journal credibility, especially in developing countries with weaker research infrastructures (Beall, 2013; Raitskaya, 2020; Narine et al., 2021; Oermann et al., 2020). While prior research has explored awareness and motivation to publish in such journals (Kharumnuid et al., 2022), limited attention has been given to academics who knowingly continue to publish in disreputable outlets. Thus, this study investigates the experiences and underlying factors influencing lecturers who, despite recognizing the lack of rigor and peer review in certain journals, continue to submit their research there. By examining their motivations and rationales, this research aims to provide deeper insights into the complex dynamics driving academics to engage with journals of questionable repute and to highlight the broader implications for scholarly publishing.

#Conclusion

The present study revealed the involvement of three Indonesian academics in publishing their work in predatory journals, driven by intense pressure to meet publication requirements for academic promotion and the difficulties of publishing in reputable, Scopus-indexed journals. This phenomenon is primarily influenced by three factors: a lack of expertise in scholarly writing for international journals, perceptions of bias and high rejection rates in top-tier publications, and the influence of religious beliefs regarding fate in academic outcomes. Publishing in predatory journals carries serious consequences for both individual academics and the broader academic community, as it can damage personal credibility, hinder career advancement, and produce research lacking scientific validity due to the absence of rigorous peer review. Consequently, it undermines institutional reputations and diminishes the overall quality and credibility of academic output, leading to a broader erosion of research integrity and the spread of unreliable knowledge. To mitigate these issues, academic institutions and policymakers must take proactive measures, such as implementing training programs on scholarly writing and predatory journal awareness, establishing mentorship initiatives, and providing workshops on international publishing standards. Furthermore, offering financial assistance for publication in reputable open-access journals, fostering international collaborations to integrate developing-country researchers into global networks, and creating frameworks to identify and blacklist predatory publishers are essential steps toward safeguarding academic integrity and promoting ethical research practices.

```python
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

text = """
Research on poor-quality journals seems to be an inexhaustible study.
The present study aims to understand the experiences of three Indonesian academics
who published their scientific articles in disreputable journals.
This phenomenology study was conducted within the framework of a qualitative approach.
Data were collected by in-depth interviews.
The study identified three key themes: Lack of Necessary Skills, Overestimation of Western Hegemony, and Acceptance of Fate.
The findings indicated that some factors caused the academics to take shortcuts to get their articles published.
The pressure to acquire promotion to become a professor was the main factor influencing Indonesian academics to publish in these fast-turnaround journals.
The findings highlight the need for academic institutions and policymakers to critically address the systemic pressures that drive scholars toward low-quality publishing outlets.
Strengthening mentorship, providing training in academic writing, and reforming promotion policies to reward quality over quantity are essential to safeguard academic integrity.

In today’s academic landscape, the expectation to publish in esteemed journals presents a significant challenge for scholars
worldwide, as universities striving for high international rankings require academics to publish in reputable international
journals (Gautam, 2020). Reputable journals are essential for fulfilling promotion criteria (Shrestha, 2021; Demir, 2018; Hedding, 2019),
and publishing original research, review articles, or thought pieces in peer-reviewed outlets demonstrates expertise and contributes
meaningfully to the field (Beall, 2012a). However, due to immense publication pressure, many scholars resort to submitting their work
to international journals despite high fees (Raitskaya, 2020; Linacre et al., 2019), often falling victim to disreputable
journals—publications that falsely present themselves as legitimate but engage in unethical practices such as dishonest editorial boards,
improper peer review, and hidden charges (Nnodim et al., 2023; Grudniewicz et al., 2019; Elmore et al., 2020). These journals exploit
researchers’ urgency to publish, leading to financial losses and damaged reputations (Beall, 2015; Cobey et al., 2018; Moher et al., 2017).
Indonesia ranks second globally in publishing within poor-quality journals (Macháček et al., 2021), driven by promotion pressure,
limited awareness, and the ease of online publishing (Xia, 2015), prompting universities and policymakers to implement awareness
campaigns, blacklists, and quality-oriented publication guidelines (Arthur et al., 2021). Studies show that lack of mentorship
and research experience also contribute to this issue, as inexperienced authors may struggle to assess journal credibility, especially
in developing countries with weaker research infrastructures (Beall, 2013; Raitskaya, 2020; Narine et al., 2021; Oermann et al., 2020).
While prior research has explored awareness and motivation to publish in such journals (Kharumnuid et al., 2022), limited attention has
been given to academics who knowingly continue to publish in disreputable outlets. Thus, this study investigates the experiences and
underlying factors influencing lecturers who, despite recognizing the lack of rigor and peer review in certain journals, continue to
submit their research there. By examining their motivations and rationales, this research aims to provide deeper insights into the
complex dynamics driving academics to engage with journals of questionable repute and to highlight the broader implications for
scholarly publishing.

The present study revealed the involvement of three Indonesian academics in publishing their work in predatory journals, driven
by intense pressure to meet publication requirements for academic promotion and the difficulties of publishing in reputable,
Scopus-indexed journals. This phenomenon is primarily influenced by three factors: a lack of expertise in scholarly writing for
international journals, perceptions of bias and high rejection rates in top-tier publications, and the influence of religious
beliefs regarding fate in academic outcomes. Publishing in predatory journals carries serious consequences for both individual
academics and the broader academic community, as it can damage personal credibility, hinder career advancement, and produce
research lacking scientific validity due to the absence of rigorous peer review. Consequently, it undermines institutional
reputations and diminishes the overall quality and credibility of academic output, leading to a broader erosion of research
integrity and the spread of unreliable knowledge. To mitigate these issues, academic institutions and policymakers must take
proactive measures, such as implementing training programs on scholarly writing and predatory journal awareness, establishing
mentorship initiatives, and providing workshops on international publishing standards. Furthermore, offering financial
assistance for publication in reputable open-access journals, fostering international collaborations to integrate developing-country
researchers into global networks, and creating frameworks to identify and blacklist predatory publishers are essential steps toward
safeguarding academic integrity and promoting ethical research practices.
"""

def words_from_text(text):
    """Ambil semua kata dalam teks"""
    return re.findall(r"[a-zA-Z]+", text.lower())

def count_vowels_consonants(word):
    """Hitung jumlah vokal dan konsonan"""
    vowels = set("aeiou")
    v = sum(1 for c in word if c in vowels)
    c = sum(1 for c in word if c.isalpha() and c not in vowels)
    return c, v

words = words_from_text(text)
unique_words = sorted(set(words))


freq_counter = Counter()
unique_counter = Counter()

for w in words:
    c, v = count_vowels_consonants(w)
    freq_counter[(c, v)] += 1

for w in unique_words:
    c, v = count_vowels_consonants(w)
    unique_counter[(c, v)] += 1


def counter_to_df(counter):
    max_c = max(k[0] for k in counter.keys())
    max_v = max(k[1] for k in counter.keys())
    df = pd.DataFrame(0, index=range(max_v+1), columns=range(max_c+1))
    for (c, v), val in counter.items():
        df.loc[v, c] = val
    return df

df_freq = counter_to_df(freq_counter)
df_unique = counter_to_df(unique_counter)

print("=== Pivot Frekuensi Kata (termasuk duplikat) ===")
print(df_freq)
print("\n=== Pivot Kata Unik (tanpa duplikat) ===")
print(df_unique)

plt.figure(figsize=(8,6))
sns.heatmap(df_freq, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Heatmap Frekuensi Kata (Jumlah Vokal vs Konsonan)")
plt.xlabel("Jumlah Konsonan")
plt.ylabel("Jumlah Vokal")
plt.show()

plt.figure(figsize=(8,6))
sns.heatmap(df_unique, annot=True, fmt=".0f", cmap="OrRd")
plt.title("Heatmap Kata Unik (Jumlah Vokal vs Konsonan)")
plt.xlabel("Jumlah Konsonan")
plt.ylabel("Jumlah Vokal")
plt.show()
```

~Kombinasi paling sering muncul ada di sekitar 2–3 konsonan dan 1–3 vokal, misalnya kata-kata pendek seperti the, and, of, in.

~Area kiri bawah (biru tua) menunjukkan kata yang paling umum dan pendek — tipikal dalam teks ilmiah berbahasa Inggris.

~Semakin ke kanan atau ke atas (lebih banyak konsonan/vokal), frekuensinya makin jarang. Ini menandakan sebagian besar kata dalam artikel ilmiah cenderung pendek dan efisien secara fonetik.

2. Heatmap Kata Unik (Tanpa Duplikat)

Visualisasi kedua menghitung hanya kata-kata unik, jadi setiap kata dihitung satu kali meskipun muncul berulang kali.

Nah pada heatmap ini kita dapat mengetahui bahwa:

~Pola distribusinya tetap mirip, tapi warna lebih lembut dan persebarannya lebih merata.

~Puncak konsentrasi ada di 3–5 konsonan dan 2–4 vokal, yang mencerminkan struktur umum kata ilmiah (study, journal, policy).

```python
import re
import matplotlib.pyplot as plt

text = """
Research on poor-quality journals seems to be an inexhaustible study.
The present study aims to understand the experiences of three Indonesian academics
who published their scientific articles in disreputable journals.
This phenomenology study was conducted within the framework of a qualitative approach.
Data were collected by in-depth interviews.
The study identified three key themes: Lack of Necessary Skills, Overestimation of Western Hegemony, and Acceptance of Fate.
The findings indicated that some factors caused the academics to take shortcuts to get their articles published.
The pressure to acquire promotion to become a professor was the main factor influencing Indonesian academics to publish in these fast-turnaround journals.
The findings highlight the need for academic institutions and policymakers to critically address the systemic pressures that drive scholars toward low-quality publishing outlets.
Strengthening mentorship, providing training in academic writing, and reforming promotion policies to reward quality over quantity are essential to safeguard academic integrity.


In today’s academic landscape, the expectation to publish in esteemed journals presents a significant challenge for scholars
worldwide, as universities striving for high international rankings require academics to publish in reputable international
journals (Gautam, 2020). Reputable journals are essential for fulfilling promotion criteria (Shrestha, 2021; Demir, 2018; Hedding, 2019),
and publishing original research, review articles, or thought pieces in peer-reviewed outlets demonstrates expertise and contributes
meaningfully to the field (Beall, 2012a). However, due to immense publication pressure, many scholars resort to submitting their work
to international journals despite high fees (Raitskaya, 2020; Linacre et al., 2019), often falling victim to disreputable
journals—publications that falsely present themselves as legitimate but engage in unethical practices such as dishonest editorial boards,
improper peer review, and hidden charges (Nnodim et al., 2023; Grudniewicz et al., 2019; Elmore et al., 2020). These journals exploit
researchers’ urgency to publish, leading to financial losses and damaged reputations (Beall, 2015; Cobey et al., 2018; Moher et al., 2017).
Indonesia ranks second globally in publishing within poor-quality journals (Macháček et al., 2021), driven by promotion pressure,
limited awareness, and the ease of online publishing (Xia, 2015), prompting universities and policymakers to implement awareness
campaigns, blacklists, and quality-oriented publication guidelines (Arthur et al., 2021). Studies show that lack of mentorship
and research experience also contribute to this issue, as inexperienced authors may struggle to assess journal credibility, especially
in developing countries with weaker research infrastructures (Beall, 2013; Raitskaya, 2020; Narine et al., 2021; Oermann et al., 2020).
While prior research has explored awareness and motivation to publish in such journals (Kharumnuid et al., 2022), limited attention has
been given to academics who knowingly continue to publish in disreputable outlets. Thus, this study investigates the experiences and
underlying factors influencing lecturers who, despite recognizing the lack of rigor and peer review in certain journals, continue to
submit their research there. By examining their motivations and rationales, this research aims to provide deeper insights into the
complex dynamics driving academics to engage with journals of questionable repute and to highlight the broader implications for
scholarly publishing.


The present study revealed the involvement of three Indonesian academics in publishing their work in predatory journals, driven
by intense pressure to meet publication requirements for academic promotion and the difficulties of publishing in reputable,
Scopus-indexed journals. This phenomenon is primarily influenced by three factors: a lack of expertise in scholarly writing for
international journals, perceptions of bias and high rejection rates in top-tier publications, and the influence of religious
beliefs regarding fate in academic outcomes. Publishing in predatory journals carries serious consequences for both individual
academics and the broader academic community, as it can damage personal credibility, hinder career advancement, and produce
research lacking scientific validity due to the absence of rigorous peer review. Consequently, it undermines institutional
reputations and diminishes the overall quality and credibility of academic output, leading to a broader erosion of research
integrity and the spread of unreliable knowledge. To mitigate these issues, academic institutions and policymakers must take
proactive measures, such as implementing training programs on scholarly writing and predatory journal awareness, establishing
mentorship initiatives, and providing workshops on international publishing standards. Furthermore, offering financial
assistance for publication in reputable open-access journals, fostering international collaborations to integrate developing-country
researchers into global networks, and creating frameworks to identify and blacklist predatory publishers are essential steps toward
safeguarding academic integrity and promoting ethical research practices.
"""

clean_text = re.sub(r'\n(?!\n)', ' ', text.strip())
paragraphs = [p.strip() for p in clean_text.split('\n\n') if p.strip()]

abbrev = ['Mr.', 'Ms.', 'Dr.', 'Prof.', 'et al.', 'No.', 'Fig.', 'Eq.', 'Ref.', 'p.']
raw_sentences = []
sentence_pattern = re.compile(r'(?<=[.!?])\s+(?=[A-Z])')

for para in paragraphs:
    temp = sentence_pattern.split(para)
    sentences = []
    for s in temp:
        if sentences and any(sentences[-1].endswith(ab) for ab in abbrev):

            sentences[-1] += ' ' + s
        else:
            sentences.append(s)
    raw_sentences.append(sentences)

sentence_counts = [len(s_list) for s_list in raw_sentences]

for i, count in enumerate(sentence_counts, 1):
    print(f"Paragraf {i} memiliki {count} kalimat.")

plt.bar(range(1, len(sentence_counts)+1), sentence_counts, color='pink', edgecolor='black')
plt.xlabel("Nomor Paragraf")
plt.ylabel("Jumlah Kalimat")
plt.title("Jumlah Kalimat dalam Tiap Paragraf (Akurasi Diperbaiki)")
plt.xticks(range(1, len(sentence_counts)+1))
plt.show()
```
